```python
import numpy as np

def O(M, D, n, m, delta):
    return ((M ** 2 * D ** 2 * n * np.log(m) * np.log(n / delta)) / m) ** 0.5

M = 1
D = 10
n = 1000
eps = 10 ** (-3)
delta = 1 - 0.999

for m in (10 ** 8, 10 ** 10, 10 ** 12, 10 ** 14):
    print(f'm = {m}: {O(M, D, n, m, delta) < eps}')
```
